```python
import pandas as pd
```

```python
rdiff = pd.read_csv('rdiff.csv', delimiter=";")
```

```python
rdiff['ts'] = pd.to_datetime(rdiff['ts'], utc=True)
```

```python
# Отобразим статистику по метрике
print("Медиана:", rdiff['value'].median())
print("Среднее:", rdiff['value'].mean())
print("Максимум:", rdiff['value'].max())
print("Минимум:", rdiff['value'].min())
print("Стандартное отклонение:", rdiff['value'].std())
```

```python
# Построим график используя Matplotlib, где на оси X будет время, а на оси Y - значение метрики

import matplotlib.pyplot as plt

# Преобразуем transaction_time в формат datetime
plt.figure(figsize=(15, 5))
plt.plot(rdiff['ts'], rdiff['value'])
plt.xlabel('Time (UTC)')
plt.ylabel('Value')
plt.title('Rdiff Metric')

plt.show()
```
