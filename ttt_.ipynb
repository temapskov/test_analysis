```python
import pandas as pd
```

```python
ttt = pd.read_csv('ttt.csv', delimiter=";")
```

```python
ttt['ts'] = pd.to_datetime(ttt['ts'], unit='us', utc=True)
```

```python
# Отобразим статистику по метрике
print("Медиана:", ttt['value'].median())
print("Среднее:", ttt['value'].mean())
print("Максимум:", ttt['value'].max())
print("Минимум:", ttt['value'].min())
print("Стандартное отклонение:", ttt['value'].std())
```

```python
# Построим график используя Matplotlib, где на оси X будет время, а на оси Y - значение метрики

import matplotlib.pyplot as plt

# Преобразуем transaction_time в формат datetime
plt.figure(figsize=(15, 5))
plt.plot(ttt['ts'], ttt['value'])
plt.xlabel('Time (UTC)')
plt.ylabel('Value')
plt.title('ttt Metric')

plt.show()
```
